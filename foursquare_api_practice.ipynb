```python

params['v'] = '20180323'

params['ll'] = ' 47.6062, -122.3321'

params['query'] = 'Starbucks'
params['radius'] = 10000
params['limit'] = 1000
response = requests.get(url=url, params=params)
data = json.loads(response.text)
```
